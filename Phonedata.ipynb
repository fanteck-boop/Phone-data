```python
plt.figure(figsize=(10, 3))

time = df["time"]
x = df["x"]
y = df["y"]
z = df["z"]

plt.plot(time, x ,color='r', label='X')  # Add label for legend
plt.plot(time, y ,color='g', label='Y')  # Add label for legend
plt.plot(time, z ,color='b', label='Z')  # Add label for legend

plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Plot of X, Y, and Z vs Time")

plt.legend()  # Improved legend formatting

plt.grid(True)
plt.xlim(0, 10)  # Set x-axis limits for 10 units
plt.ylim(-1.5, 1.5)
plt.tight_layout()
plt.show()
```
